```python
#Clean fake user names
data=data.drop_duplicates()
data=data[~data['User_Name'].str.match(r'^\d+$',na=False)]
data=data[data['User_Name'].str.len()>3]
data
```

```python
#Date Format correction
data['Posted_Date']=pd.to_datetime(data['Posted_Date'],errors='coerce')
data
```

```python
data['Posted_Date'] = data['Posted_Date'].dt.strftime('%Y-%m-%d')
data
```

```python
#Handling Shares column missing data
data['Shares']=data['Shares'].fillna(0)
data
```
